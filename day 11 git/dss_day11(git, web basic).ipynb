git 기본 명령어 

 - git init
  + 새로운 깃 저장소 추가
  
 - git remote origin add [repository address]
  + 리모트의 리포지토리를 로컬과 연결
  
 - git add [file]
  + 스테이지(준비영역)에 변경 이력 커밋 준비
  
 - git commit -m "message"
  + HEAD(로컬저장소)에 변경이력 추가
  
 - git push origin [branch]
  + 원격 저장소에 추가
  
 - git pull (git fetch + git merge)
  + 원격 저장소에서 최신 이력을 받아와, 로컬저장소와 작업공간의 3 way merge
 
 - git fetch + git rebase
  + 원격 저장소에서 최신 이력을 받아와, 로컬저장소에 rebase 

```python
# rebase 
git fetch origin master
git rebase origin/master
(수동으로 충돌나는 파일 머지) if 충돌발생시
git add .
git rebase --continue
git push origin master

# merge
git pull origin master
 - git fetch origin master
 - git merge origin/master

(수동으로 충돌나는 파일 머지) if 충돌발생시
git push origin master
```

 * 연습문제)
  - 주소를 입력받아 해당 위치의 좌표를 전송하는 함수를 만들어서 각자 집의 좌표를 확인해보세요.
  - 집에서 신사역까지의 직선 거리를 구해보세요.
  

```python
from math import sin, cos, sqrt, atan2, radians
API_KEY = ''

def get_geocode(address):
    try:
        api_end_point = 'https://maps.googleapis.com/maps/api/geocode/json?address={}&key={}'.format(address, API_KEY)

        res = requests.get(api_end_point)
        data = json.loads(res.text) # loads method이용

        latlng = data['results'][0]['geometry']['location']
        return latlng['lat'], latlng['lng']
    except Exception as e:
        return 37.11212, 127.21312
    
    
lat1, lng1 = get_geocode('집')
lat2, lang2 = get_geocode('신사')

# 두개의 위경도가 주어졌을 때, 거리 구하는 함수
def get_distance(lat1, lng1, lat2, lng2):
    R = 6373.0

    lat1 = radians(lat1)
    lng1 = radians(lng1)
    lat2 = radians(lat2)
    lng2 = radians(lng2)

    dlon = lng2 - lng1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

* bs4 (BeautifulSoup) module
 - 가장 널리 쓰이는 html parsing 모듈
 - html 각 요소의 값을 다양한 방법으로 검색가능
 - https://www.crummy.com/software/BeautifulSoup/bs4/doc/

* Web page html 파헤치기
 - chrome, IE 등 브라우져의 개발자 도구를 이용하여 분석

```python
import requests
from bs4 import BeautifulSoup

def get_news_title(url):
    response = requests.get(url)
    content = response.text
    
    soup = BeautifulSoup(content)
    title = soup.title.get_text()
    
    return title


print get_news_title('http://media.daum.net/foreign/newsview?newsid=20160921114543616')


def get_news_content(url):
    response = requests.get(url)
    content = response.text

    soup = BeautifulSoup(content)

    div = soup.find('div', attrs = {'id' : 'harmonyContainer'})
    
    soup.find_all('p')
    
    content = ''
    for paragraph in div.find_all('p'):
        content += paragraph.get_text()
        
    return content.encode('utf-8')
        
news1 = get_news_content('http://v.media.daum.net/v/20170124105051203')
print news1
```

```python
def get_naver_news_title(url):
    response = requests.get(url)
    content = response.text
    soup = BeautifulSoup(content)
    
    #soup.title.get_text()
    h3 = soup.find('h3', attrs = {'id' : 'articleTitle'})
    return h3.get_text()

def get_naver_news_body(url):
    try:
        response = requests.get(url)
        content = response.text
    except:
        return ''
    
    soup = BeautifulSoup(content)
    
    div = soup.find('div', attrs = {'id' : 'articleBodyContents'})
    if div == None:
        return ''
    
    return div.get_text().strip()

print get_naver_news_title('http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=001&aid=0008982535')
print get_naver_news_body('http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=001&aid=0008982535')
```
