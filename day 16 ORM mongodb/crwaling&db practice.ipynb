```python
import os
import sys
from sqlalchemy import Column, ForeignKey, Integer, CHAR, Date, String, Time, Index, DateTime, TIMESTAMP, func
from sqlalchemy.dialects.mysql import INTEGER, BIT, TINYINT, TIME, DOUBLE, TEXT
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship
from sqlalchemy import create_engine
from sqlalchemy import PrimaryKeyConstraint
from sqlalchemy.orm import sessionmaker
import requests
from bs4 import BeautifulSoup
import datetime

Base = declarative_base()

class News(Base):
    __tablename__ = 'news'
    
    ID = Column(Integer, primary_key = True, nullable = False, autoincrement = True)
    Title = Column(CHAR(50), nullable = False)
    Content = Column(CHAR(5000), nullable = False)
    CrawlingTime = Column(CHAR(20), nullable = False)
    
```

```python
engine = create_engine(connection_string, pool_recycle = 3600,encoding='utf-8')
Session = sessionmaker(bind=engine)

session = Session()
```

```python
#base_url = "http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105"
#url = base_url.format()
url = "http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=001&aid=0009076202"
res = requests.get(url)
soup = BeautifulSoup(res.text)

title = soup.select('title')[0].get_text()
body = soup.select("div#articleBodyContents")[0].get_text()
now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
now = now.replace("-"," ")
now = now.replace(":"," ")
now

new = News(Title = title, Content = body, CrawlingTime = now)
session.add(new)
session.commit()
```
