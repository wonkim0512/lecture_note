* MySQL-python module 실습

```python
# mysql server가 설치된 호스트 주소
server = 'ec2-35-167-74-63.us-west-2.compute.amazonaws.com' 
```

* select one

```python
import MySQLdb
# 접속할 서버 주소, DB 사용자 아이디, DB 사용자 비밀번호, DB
db = MySQLdb.connect(server, 'root','root','world' ) 

cursor = db.cursor()

# 쿼리를 문자열로 전닯
cursor.execute("select * from country")

# 한개만 가져옴
data = cursor.fetchone()

print type(data)
print data #tuple로 반환

# 연결 종료
db.close()
```

* select all

```python
import MySQLdb
db = MySQLdb.connect(server, 'root','root','world' ) 

cursor = db.cursor()
cursor.execute("select * from country")

print cursor.rowcount

# 모두 가져옴
results = cursor.fetchall()
for row in results:
    print row

# disconnect from server
db.close()
```

* insert data

```python
import MySQLdb
db = MySQLdb.connect(server, '',"","", cursorclass=MySQLdb.cursors.DictCursor) #커서타입 지정가능.

cursor = db.cursor()

try:
   # transaction
   cursor.execute("""INSERT INTO city VALUES (10001,'Pangyo','KOR','PAN',170000);""") # 코드에서는 ;(세미 콜론) 굳이 안넣어줘도 됨
   cursor.execute("""INSERT INTO city VALUES (10002,'Pangyo','KOR','PAN',170000);""")
   cursor.execute("""INSERT INTO city VALUES (10003,'Pangyo','KOR','PAN',170000);""")
   cursor.execute("""INSERT INTO city VALUES (10004,'Pangyo','KOR','PAN',170000);""")
        
   db.commit() #워크벤치는 바로 commit을 시켜주는 거고, 파이썬에서는 우리가 직접 해야함.
except Exception as e:
   db.rollback()
   print e


# disconnect from server
db.close()
```

- transaction의 개념이 가장 중요하다.
- 모두가 successful해야 진행됨. 
- 위로 예를 들면, 네번째 execute에서 에러가 뜨면 위의 3개도 모두 취소됨.
